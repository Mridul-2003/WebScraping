```python
import requests
from bs4 import BeautifulSoup
url = "https://gov.optimism.io"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
categories = soup.find_all("td",class_="category")
```

```python
categories
```

```python
name=[]
for headings in categories:
  heading = headings.find("h3").text
  name.append(heading)
```

```python
name
```

```python
names = [item.strip() for item in name]
```

```python
names
```

```python
category_links = []
for links in categories:
  link = links.find("a")["href"]
  category_links.append("https://gov.optimism.io"+link)
category_links
```

```python
category_description = []
for descriptions in categories:
  description = descriptions.find('div', {'itemprop': 'description'}).text
  category_description.append(description)
category_description
```

```python
import pandas as pd
df = pd.DataFrame({'Category_Name':names,'Category_link':category_links,'Category_Description':category_description})
df.head()
```
