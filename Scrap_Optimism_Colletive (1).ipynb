```python
import requests
from bs4 import BeautifulSoup

category_links = []
# Assuming categories is defined earlier in your code
for links in categories:
    link = links.find("a")["href"]
    category_links.append("https://gov.optimism.io" + link)

# To store the article names with URL and heading links
article_names = []

# Function to scrape a single page for articles
def scrape_articles(page_url):
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Find all article elements
    articles = soup.find_all("td", class_="main-link")

    # Extract article headings and their links
    new_articles = []
    for headings in articles:
        heading = headings.find("a").text.strip()
        heading_link = headings.find("a")["href"]

        # Append each article's details as a dictionary to the list
        new_articles.append({
            "Heading": heading,
            "Heading_link": heading_link
        })

    return new_articles, soup

# Function to handle pagination with ?page=
def scrape_category_with_pagination(category_url):
    page_num = 0
    latest_headings = set()  # To track latest headings and avoid duplicates
    has_more_pages = True

    while has_more_pages:
        # Construct the URL with pagination
        current_page_url = f"{category_url}?page={page_num}"
        new_articles, soup = scrape_articles(current_page_url)

        # Check if new articles are indeed new (to avoid duplicates)
        if not new_articles:  # Stop if no articles are found on this page
            has_more_pages = False
            break

        for article in new_articles:
            if article["Heading"] not in latest_headings:
                article_names.append({
                    "URL": current_page_url,  # Add the URL of the current page
                    "Heading": article["Heading"],
                    "Heading_link": article["Heading_link"]
                })
                latest_headings.add(article["Heading"])

        # Increment page number for the next loop iteration
        print("Page Url = ",current_page_url)
        print("Page Number = ",page_num)
        page_num += 1

# Scrape articles for each category with pagination
for category_link in category_links:
    scrape_category_with_pagination(category_link)

# This will now give you a list of dictionaries with each article's URL and heading
print(article_names)
```
